```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib

# read files
out1 = np.genfromtxt('pos_spt.out',comments='%')
out2 = np.genfromtxt('pos_gpt.out',comments='%')
out3 = np.genfromtxt('pos_gpt2.out',comments='%')

# plot
plt.figure(figsize=(10,5))
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=False)
ax1.plot(out1[:,2],'bo')
ax1.plot(out2[:,2],'ro',alpha=0.4)
ax1.plot(out3[:,2],'go',alpha=0.4)
ax2.plot(out1[:,3],'bo')
ax2.plot(out2[:,3],'ro')
ax3.plot(out1[:,4],'bo')
ax3.plot(out2[:,4],'ro')
f.subplots_adjust(hspace=0)
f.set_figheight(15)
f.set_figwidth(15)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
plt.show()
```
